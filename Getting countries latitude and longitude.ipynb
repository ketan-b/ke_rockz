```python
import bs4
import requests
import pandas as pd
```

```python
result = requests.get("https://developers.google.com/public-data/docs/canonical/countries_csv")
soup = bs4.BeautifulSoup(result.content)
```

```python
table = soup.find('table')
```

```python
rows = []
columns = ["Country", "Latitude", "Longitude", "Name"]
for tr in table.find_all('tr'):
    row = []
    for td in tr.find_all('td'):
        row.append(td.get_text().strip())
    if len(row) > 0:
        rows.append(pd.DataFrame([row], columns=columns))
df = pd.concat(rows, ignore_index=True)
```

```python
df.set_index("Name", inplace = True)
df.drop(["Country"], axis=1, inplace=True)
```

```python
df.to_json('./datasets/countries.csv')
```
