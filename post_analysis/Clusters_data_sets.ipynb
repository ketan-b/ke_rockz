```python
import pandas as pd
```

```python
export_commodities_by_air = pd.read_json('ea.json')
export_commodities_by_sea = pd.read_json('es.json')
import_commodities_by_sea = pd.read_json('is.json')
import_commodities_by_air = pd.read_json('ia.json')
```

```python
export_commodities_by_air.drop(['Latitude', 'Longitude'], axis=1, inplace=True)
export_commodities_by_sea.drop(['Latitude', 'Longitude'], axis=1, inplace=True)
import_commodities_by_sea.drop(['Latitude', 'Longitude'], axis=1, inplace=True)
import_commodities_by_air.drop(['Latitude', 'Longitude'], axis=1, inplace=True)
```

```python
summed_ea = export_commodities_by_air.groupby('Cluster Labels').sum()
list_ea = export_commodities_by_air.groupby('Cluster Labels').apply(lambda x: ', '.join(x.index.tolist()))
ea_desc = summed_ea.merge(list_ea.rename('Countries'), left_index=True, right_index=True)
```

```python
summed_es = export_commodities_by_sea.groupby('Cluster Labels').sum()
list_es = export_commodities_by_sea.groupby('Cluster Labels').apply(lambda x: ', '.join(x.index.tolist()))
es_desc = summed_es.merge(list_es.rename('Countries'), left_index=True, right_index=True)
```

```python
summed_is = import_commodities_by_sea.groupby('Cluster Labels').sum()
list_is = import_commodities_by_sea.groupby('Cluster Labels').apply(lambda x: ', '.join(x.index.tolist()))
is_desc = summed_is.merge(list_is.rename('Countries'), left_index=True, right_index=True)
```

```python
summed_ia = import_commodities_by_air.groupby('Cluster Labels').sum()
list_ia = import_commodities_by_air.groupby('Cluster Labels').apply(lambda x: ', '.join(x.index.tolist()))
ia_desc = summed_ia.merge(list_ia.rename('Countries'), left_index=True, right_index=True)
```
